```python
import torch
import torchaudio
from IPython.display import Audio
```

```python
# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# ✅ Fix: Unpack only what is returned
model, example_text = torch.hub.load(
    repo_or_dir="snakers4/silero-models",
    model="silero_tts",
    language="en",
    speaker="v3_en",
    trust_repo=True
)

# Set the correct sample rate manually (Silero TTS uses 48kHz)
sample_rate = 48000
```

```python
# Move model to GPU if available
model.to(device)
```

```python
# List of different voices
speakers = ["en_0", "en_55", "en_10", "en_20", "en_78"]  # You can change these speaker IDs
```

```python
# Function to generate speech in multiple voices
def generate_speech(text, speakers, save_dir="output_files"):
    import os
    os.makedirs(save_dir, exist_ok=True)  # Create directory if not exists

    output_paths = []  # To store file paths

    for speaker in speakers:
        audio = model.apply_tts(text=text, speaker=speaker)  # Generate audio
        audio_tensor = torch.tensor(audio).unsqueeze(0)  # Convert to tensor

        save_path = f"{save_dir}/output_{speaker}.wav"
        torchaudio.save(save_path, audio_tensor, sample_rate)
        output_paths.append(save_path)
        print(f"Saved: {save_path}")

    return output_paths
```

```python
# Example Usage
text_input = "Hello, we will learn Deep learning today "
# Generate speech in multiple voices
output_files = generate_speech(text_input, speakers)
```

```python
# Play all generated voices
for output_path in output_files:
    print(f"Playing: {output_path}")
    display(Audio(output_path))
```

```python
model2 = torch.hub.load(
    repo_or_dir="snakers4/silero-models",
    model="silero_tts",
    language="indic",  # Keeping the language as 'indic'
    speaker="v3_indic",  # You can also try 'v4_indic'
    trust_repo=True
)

# Set the sample rate manually (Silero TTS uses 48kHz)
sample_rate = 48000
```

```python
def generate_speech(text, save_path="output.wav", speaker="v3_indic"): # Added speaker argument with default value
    # Generate speech audio from text
    audio = model2.apply_tts(text=text, speaker=speaker)  #Pass speaker argument to apply_tts

    # Convert to torch tensor and reshape
    audio_tensor = torch.tensor(audio).unsqueeze(0)

    # Save as WAV file
    torchaudio.save(save_path, audio_tensor, sample_rate)

    return save_path
```
